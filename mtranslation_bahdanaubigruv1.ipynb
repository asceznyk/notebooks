```
import re

import numpy as np
import pandas as pd

import time
import torch
import spacy
import torchtext
import random
from torchtext.data import Field, BucketIterator, TabularDataset
from sklearn.model_selection import train_test_split

class tokenize(object):
    
    def __init__(self, lang):
        self.nlp = spacy.load(lang)
            
    def tokenizer(self, sentence):
        sentence = re.sub(
        r"[\*\"“”\n\\…\+\-\/\=\(\)‘•:\[\]\|’\!;]", " ", str(sentence))
        sentence = re.sub(r"[ ]+", " ", sentence)
        sentence = re.sub(r"\!+", "!", sentence)
        sentence = re.sub(r"\,+", ",", sentence)
        sentence = re.sub(r"\?+", "?", sentence)
        sentence = sentence.lower()
        return [tok.text for tok in self.nlp.tokenizer(sentence) if tok.text != " "]

en = tokenize('en')
de = tokenize('de')

SRC = Field(tokenize=de.tokenizer, init_token = '<sos>', eos_token = '<eos>')
TRG = Field(tokenize=en.tokenizer, init_token = '<sos>', eos_token = '<eos>')
```

```
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch import Tensor
```

```
ENCEMBDIM = 256
DECEMBDIM = 256
ENCHIDDIM = 128
DECHIDDIM = 128
ATTNDIM = 64
ENCDROPOUT = 0.5
DECDROPOUT = 0.5
```

```
class EncoderBiGRU(nn.Module):
    def __init__(self, 
                 input_dim: int,
                 emb_dim: int, 
                 hidden_dim: int, 
                 act_dim: int,
                 dropout_rate: float):
        super().__init__()

        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.emb_dim = emb_dim
        self.act_dim = act_dim

        self.dropout = nn.Dropout(dropout_rate)
        self.embed = nn.Embedding(self.input_dim, self.emb_dim)
        self.gru = nn.GRU(self.emb_dim, self.hidden_dim, bidirectional=True)
        self.fc = nn.Linear(self.hidden_dim * 2, self.act_dim)
               
    def forward(self, x_s):
        x_s = self.dropout(self.embed(x_s))

        a_s, h_t = self.gru(x_s)

        h_t = torch.cat((h_t[-2,:,:], h_t[-1,:,:]), dim=1) # why -2, -1 ? 
        h_t = torch.tanh(self.fc(h_t))
        
        return (a_s, h_t)
```

```
class BahdanauAttention(nn.Module):
    def __init__(self, act_dim, hidden_dim, attn_dim):
        super().__init__()
        self.act_dim = act_dim
        self.hidden_dim = hidden_dim
        self.attn_dim = attn_dim

        self.w_e = nn.Linear(self.hidden_dim, self.attn_dim)
        self.u_e = nn.Linear(2 * self.act_dim, self.attn_dim)
        self.v_e = nn.Linear(self.attn_dim, 1)

    #h_t.shape = (batch_size, hidden_dim) to (max_len, batch_size, hidden_dim)
    #a_s.shape = (max_len, batch_size, 2 * act_dim)

    def forward(self, h_t, a_s):
        z_e1 = self.u_e(a_s)

        h_t = h_t.repeat(a_s.shape[0], 1, 1) #h_t shape is now (max_len, batch_size, hidden_dim)
        z_e2 = self.w_e(h_t)

        e_t = self.v_e(torch.tanh(z_e1 + z_e2)) #e_t.shape should be ... (max_len, batch_size, 1)
        e_t = torch.squeeze(e_t, dim=2).T

        return F.softmax(e_t, dim=1) #this shape is ... (batch_size, max_len)
```

```
class DecoderContextGRU(nn.Module):
    def __init__(self, act_dim, emb_dim, hidden_dim, output_dim, dropout_rate, attention):
        super().__init__() 
        self.act_dim = act_dim
        self.emb_dim = emb_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.attention = attention
        
        self.dropout = nn.Dropout(dropout_rate)
        self.embed = nn.Embedding(self.output_dim, self.emb_dim)
        self.gru = nn.GRU((self.act_dim) * 2 + self.emb_dim, self.hidden_dim)
        self.fc = nn.Linear(self.hidden_dim, self.output_dim)

    def calc_ctx(self, h_t, a_s):
        alpha_t = self.attention(h_t, a_s)
        alpha_t = alpha_t.unsqueeze(1) #alpha_t.shape = (batch_size, 1, max_len) 

        a_s = a_s.permute(1,0,2) #a_s.shape = (batch_size, max_len, 2 * act_dim)

        c_t = torch.bmm(alpha_t, a_s) #c_t.shape = (batch_size, 1, 2 * act_dim)

        return c_t.permute(1, 0, 2) #c_t.shape (1, batch_size, 2 * act_dim)

    def forward(self, i_t, a_s, h_t):
        i_t = i_t.unsqueeze(0) #i_t.shape = (1, batch_size, 1)
        i_t = self.embed(i_t) #i_t.shape = (1, batch_size, emb_dim)

        c_t = self.calc_ctx(h_t, a_s)

        a_t, h_t = self.gru(torch.cat((i_t, c_t), dim=2), h_t.unsqueeze(0))
        #a_t, h_t (1, batch_size, hidden_dim)

        a_t, h_t = a_t.squeeze(0), h_t.squeeze(0) 
        #a_t and h_t are squeezed to shape (batch_size, hidden_dim)     
           
        y_t = self.fc(a_t)

        return y_t, h_t #y_t, h_t (batch_size, output_dim), (batch_size, hidden_dim)
```

```
encoder = EncoderBiGRU(SRCDIM, ENCEMBDIM, ENCHIDDIM, DECHIDDIM, ENCDROPOUT)
attention = BahdanauAttention(ENCHIDDIM, DECHIDDIM, ATTNDIM)
decoder = DecoderContextGRU(ENCHIDDIM, DECEMBDIM, DECHIDDIM, TRGDIM, DECDROPOUT, attention)
```

```
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, trg, teacher_force: float = 0.5):
        batch_size = src.shape[1]
        max_len = trg.shape[0]
        x_s = src

        p_t = torch.zeros(max_len, batch_size, self.decoder.output_dim).to(self.device) #p_t stands for prediction
        y_t = trg[0, :] # the encoding for <sos> token 

        a_s, h_t = self.encoder(x_s) #h_t and a_s are encodings

        for t in range(1, max_len):
            y_t, h_t = self.decoder(y_t, a_s, h_t) #a_s is repeated due to attention calculation
            p_t[t] = y_t
            y_t = y_t.max(1)[1] #max returns two tensors the max values and args (indices)
            y_t = (trg[t, :] if teacher_force > random.random() else y_t)

        return p_t

# p_t[t].max(2)[1] will be call the args where there are max logits
# t is the index time-step where max args can be calculated
```

```
model = Seq2Seq(encoder, decoder, device).to(device)
```

```
def init_weights(m: nn.Module):
    for name, param in m.named_parameters():
        if 'weight' in name:
            nn.init.normal_(param.data, mean=0, std=0.01)
        else:
            nn.init.constant_(param.data, 0)


model.apply(init_weights)

optimizer = optim.Adam(model.parameters())


def count_parameters(model: nn.Module):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


print(f'The model has {count_parameters(model):,} trainable parameters')
```

```
PADIDX = TRG.vocab.stoi['<pad>']
criterion = nn.CrossEntropyLoss(ignore_index=PADIDX)
```

```
def train_model(model:nn.Module, criterion:nn.Module, iterator:BucketIterator, clip:float):
    model.train()

    loss_epoch = 0

    for i, batch in enumerate(iterator):
        optimizer.zero_grad()

        src = batch.German
        trg = batch.English

        outputs = model(src, trg)
        outputs = outputs.view(-1, outputs.shape[-1])
        trg = trg.view(-1)

        loss = criterion(outputs, trg)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), clip)

        optimizer.step()

        loss_epoch += loss.item()
    
    return loss_epoch / len(iterator)
```

```
def evaluate_model(model:nn.Module, criterion:nn.Module, iterator:BucketIterator):
    model.eval()

    loss_epoch = 0

    with torch.no_grad():

        for i, batch in enumerate(iterator):
            src = batch.German
            trg = batch.English

            outputs = model(src, trg, teacher_force=0)
            
            outputs = outputs.view(-1, outputs.shape[-1])
            trg = trg.view(-1)

            loss = criterion(outputs, trg)

            loss_epoch += loss.item()

    return loss_epoch / len(iterator)
```

```
def epoch_time(start_time: int, end_time: int):
    elapsed_time = end_time - start_time
    elapsed_mins = int(elapsed_time / 60)
    elapsed_secs = int(elapsed_time - (elapsed_mins * 60))
    return elapsed_mins, elapsed_secs
```

```
NEPOCHS = 10
CLIP = 1

for epoch in range(NEPOCHS):
    start_time = time.time()

    loss_train = train_model(model, criterion, train_iterator, CLIP)
    loss_val = evaluate_model(model, criterion, valid_iterator)

    end_time = time.time()

    epoch_mins, epoch_secs = epoch_time(start_time, end_time)

    print(f'Epoch: {epoch+1} Epoch Time: {epoch_mins} mins {epoch_secs} secs')

    print(f'Training Loss: {loss_train:.3f} | Training PPL: {np.exp(loss_train):.3f}')
    print(f'Validation Loss: {loss_val:.3f} | Validation PPL: {np.exp(loss_val):.3f}')
```

```
def translate_text(model: nn.Module, text: str, trglen: int):
    src = np.array([[SRC.vocab.stoi[s]] for s in text.split(' ')])
    trg = np.zeros((trglen, 1))
    trg[0, :] = SRC.vocab.stoi['<sos>'] 

    src = torch.LongTensor(src).to(device)
    trg = torch.LongTensor(trg).to(device)

    output = model(src, trg, 0)
    output = output.max(2)[1]

    return ' '.join([TRG.vocab.itos[i] for i in list(output.cpu().detach().numpy().flatten())])
```
